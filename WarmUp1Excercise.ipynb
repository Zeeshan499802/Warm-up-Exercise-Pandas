Find Total Number of Students Having Marks Between 90 To 100 (Inclusive) Using Between Method 

```python
# There are three method Used 

# First Method 
df[df['Marks']>=90]

# Second Method 
len(df[(df['Marks']>=90) & (df['Marks']<=100)])

# Third Method 
sum(df['Marks'].between(90,100))
```

Apply Method 

```python
def marks(x):
  # return x/2 #Normal Division
  return x//2 #Floating Division

df['Half_Marks'] = df['Marks'].apply(marks)
df

```

Using Lambda Function 

```python
df['Marks'].apply(lambda x:x//2)
```

Map Function 


```python
df['Male_Female']=df['Gender'].map({'Male':1,'Female':0})
df
```

Drop the Column 

```python
df.drop(['Male_Female','Half_Marks'],axis=1,inplace=True)
df
```
